```python
def ImpliedMerton(c):
    #print(A,K,ÏƒE)
    d1 = (np.log(A/K) + (r+c[1]**2/2)*T)/(c[1] * np.sqrt(T))
    d2 = d1 - c[1] * np.sqrt(T)
    #print(d1,d1)
    res1 = A* norm.cdf(d1) - np.exp(-r*T) * K * norm.cdf(d2) - c[0]
    res2 = (ÏƒE * norm.cdf(d1) * c[0]) - A*c[1]
    #print(res1**2+res2**2)
    return(res1**2+res2**2)
    #print("[V, ÏƒV]=",c,", Object Function Value:", val)

def default_prob(A,K,ÏƒA,T,r):
    dd= (np.log(A/K) + (r-(ÏƒA**2)/2)*T)/( ÏƒA* np.sqrt(T))

    return norm.cdf(-dd)

```

```python
time_series_data = default_prob_df['PFE Default Probability']

# Threshold for crossovers
threshold = df_realised_deafult['Default Prob'].iloc[0]

# Generate crossover signals based on the threshold
above_threshold_signals = [value > threshold for value in time_series_data]
below_threshold_signals = [value < threshold for value in time_series_data]

# Plotting
plt.figure(figsize=(15,20))


# Plot the time series data
plt.plot(time_series_data, label='Time Series', marker='o')

# Plot crossover signals above the threshold
plt.scatter([i for i, signal in enumerate(above_threshold_signals) if signal], [value for i, value in enumerate(time_series_data) if above_threshold_signals[i]], color='green', label='Above Threshold', marker='^', s=100)

# Plot crossover signals below the threshold
plt.scatter([i for i, signal in enumerate(below_threshold_signals) if signal], [value for i, value in enumerate(time_series_data) if below_threshold_signals[i]], color='red', label='Below Threshold', marker='v', s=100)

# Add a horizontal line at the threshold
plt.axhline(y=threshold, color='blue', linestyle='--', label='Threshold')

# Add labels and title
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Time Series with Crossover Signals')

# Add legend
plt.legend()

# Show the plot
plt.show()
```
